DADOS DESPESAS ORÇAMENTÁRIAS TOTAL

```python
import pandas as pd
import glob
import os

caminho_da_pasta = '../data/external/dataset_tce/dados_orcamentos'

pasta_de_saida = '../data/processed/'
arquivo_de_saida = 'orcamentos_tce_consolidados.csv'

caminho_completo_saida = os.path.join(pasta_de_saida, arquivo_de_saida)

# glob serve para encontrar todos os arquivos .csv na pasta
todos_os_arquivos = glob.glob(os.path.join(caminho_da_pasta, "*.csv"))

lista_de_dataframes = []


for arquivo in todos_os_arquivos:
   
    df_temp = pd.read_csv(arquivo)
    lista_de_dataframes.append(df_temp)


if lista_de_dataframes:

   # ignore o index dos arquivos
    df_final = pd.concat(lista_de_dataframes, ignore_index=True)

    df_final.to_csv(caminho_completo_saida, index=False, encoding='utf-8')
    
    print("\n--- Processo Concluído! ---")
    print(f"Os dados foram consolidados com sucesso no arquivo: '{arquivo_de_saida}'")
    print(f"O arquivo final possui {df_final.shape[0]} amostras e {df_final.shape[1]} features.")

    print("\n---- Valores Nulos por feature ----")
    print(df_final.isnull().sum())

    df_final.info()

else:
    print("Nenhum arquivo CSV foi encontrado na pasta especificada. Verifique o caminho.")
```
